```python
%%sql
SELECT district, MAX(CASE WHEN DACOITY > ROBBERY THEN DACOITY ELSE ROBBERY END) AS max_incidents
FROM ST
GROUP BY DISTRICT
ORDER BY max_incidents DESC
LIMIT 1;
```

```python
%%sql
SELECT DISTRICT, SUM(MURDER) AS total_murders
FROM ST
GROUP BY DISTRICT
HAVING total_murders = (SELECT MIN(total_murders) FROM (SELECT SUM(MURDER) AS total_murders FROM ST GROUP BY DISTRICT) AS subquery);
```

```python
%%sql
SELECT DISTRICT, YEAR, SUM(MURDER) AS total_murders
FROM ST
GROUP BY DISTRICT, YEAR
ORDER BY DISTRICT ASC, YEAR ASC;
```

```python
%%sql
SELECT 
    D.STATE, 
    D.YEAR, 
    D.DISTRICT, 
    D.MURDER
FROM 
    DISTRICT D
JOIN 
    (
        SELECT 
            STATE, 
            YEAR, 
            MAX(MURDER) AS MAX_MURDER
        FROM 
            DISTRICT
        GROUP BY 
            STATE, YEAR
    ) AS MaxMurder
ON 
    D.STATE = MaxMurder.STATE
    AND D.YEAR = MaxMurder.YEAR
    AND D.MURDER = MaxMurder.MAX_MURDER;
```

```python
import pandas as pd
import sqlite3
```

```python
df = pd.read_excel('DISTRICT.xlsx')
print(df)
```

```python
conn = sqlite3.connect(':memory:')
df.to_sql('DISTRICT', conn, index=False)
```

```python
query = """
WITH RankedMurders AS (
    SELECT
        STATE,
        DISTRICT,
        YEAR,
        MURDER,
        ROW_NUMBER() OVER (PARTITION BY STATE, YEAR ORDER BY MURDER DESC) AS MurderRank
    FROM
        DISTRICT
)
SELECT
    STATE,
    YEAR,
    DISTRICT,
    MURDER
FROM
    RankedMurders
WHERE
    MurderRank = 1;
"""
highest_murders = pd.read_sql_query(query, conn)

print("\nHighest number of murders:")
print(highest_murders)

# Close the connection
conn.close()
```

```python
df1 = highest_murders
```

```python
df1.info()
```

```python
# Count the number of appearances of each district
district_counts = df1['DISTRICT'].value_counts()

# Filter districts that appear 3 or more than 3 years
selected_districts = district_counts[district_counts >= 3].index

# Filter the DataFrame based on selected districts
filtered_df = df1[df1['DISTRICT'].isin(selected_districts)]

# Sort the filtered DataFrame by district, year, and murders in descending order
sorted_df = filtered_df.sort_values(by=['DISTRICT', 'YEAR', 'MURDER'], ascending=[True, False, False])

# Print the sorted DataFrame
print(sorted_df[['STATE', 'DISTRICT', 'MURDER', 'YEAR']])
```

```python
sorted_df['DISTRICT'].value_counts()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Line Plot: Trend of murders over the years for each district
plt.figure(figsize=(10, 6))
sns.lineplot(data=sorted_df, x='YEAR', y='MURDER', hue='DISTRICT', marker='o')
plt.title('Trend of Murders Over Years for Each District')
plt.xlabel('Year')
plt.ylabel('Number of Murders')
plt.legend(title='District')
plt.show()

# Bar Plot: Total number of murders in each district
plt.figure(figsize=(10, 6))
sns.barplot(data=sorted_df, x='DISTRICT', y='MURDER', estimator=sum)
plt.title('Total Number of Murders in Each District')
plt.xlabel('District')
plt.ylabel('Total Number of Murders')
plt.xticks(rotation=45)
plt.show()

# Stacked Bar Plot: Distribution of murders by year for each district
plt.figure(figsize=(10, 6))
sns.barplot(data=sorted_df, x='DISTRICT', y='MURDER', hue='YEAR', estimator=sum)
plt.title('Distribution of Murders by Year for Each District')
plt.xlabel('District')
plt.ylabel('Total Number of Murders')
plt.xticks(rotation=45)
plt.legend(title='Year')
plt.show()

# Heatmap: Intensity of murders across districts and years
pivot_df = sorted_df.pivot_table(values='MURDER', index='DISTRICT', columns='YEAR', aggfunc=sum)
plt.figure(figsize=(10, 6))
sns.heatmap(pivot_df, cmap='Reds')
plt.title('Intensity of Murders Across Districts and Years')
plt.xlabel('Year')
plt.ylabel('District')
plt.show()
```
